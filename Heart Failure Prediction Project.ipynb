# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']
print('Shape of x =', x.shape)
print('Shape of y =', y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=70)
print('Shape of x_train = ', x_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of x_test = ', x_test.shape)
print('Shape of y_test = ', y_test.shape)
```

```python
sc = StandardScaler()
sc.fit(x_train)
```

# Support Vector Calssification

```python
from sklearn.svm import SVC
```

```python
classification_rbf = SVC(kernel="rbf")
classification_rbf.fit(x_train,y_train)
```

```python
classification_rbf.score(x_test,y_test)
```

```python
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
classification_rbf_2=SVC(kernel='rbf')
classification_rbf_2.fit(x_train_sc,y_train)
classification_rbf_2.score(x_test_sc,y_test)
```

# Tune Parameter

```python
model_c=SVC(C=5)
model_c.fit(x_train,y_train)
model_c.score(x_test,y_test)
```

```python
model_g=SVC(gamma=5)
model_g.fit(x_train,y_train)
model_g.score(x_test,y_test)
```

```python
classification_poly=SVC(kernel='poly',degree=5)
classification_poly.fit(x_train_sc,y_train)
classification_poly.score(x_test_sc,y_test)
```

```python
classification_linear=SVC(kernel='linear',degree=5)
classification_linear.fit(x_train_sc,y_train)
classification_linear.score(x_test_sc,y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm,annot=True,fmt='g')
plt.ylabel('age',fontsize=13)
plt.xlabel('anaemia',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
precision=precision_score(y_test,y_pred,average='macro')
print(precision)
```

```python
recall=recall_score(y_test,y_pred,average='macro')
print(recall)
```

```python
f1_score=f1_score(y_test,y_pred,average='macro')
print(f1_score)
```
